# Postal Code Information

```python
response = requests.get("http://api.zippopotam.us/us/ma/belmont")
zipcode = json.loads(response.text)
#zipcode
```

```python
from IPython.display import Image
Image(filename='img/img1.png', width=300, height=300)
```

```python
for item in zipcode:
    print(item)
```

```python
zipcode['country abbreviation']
```

```python
zipcode['places']
```

```python
name = []
post = []
longitude = []
latitude = []
for place in zipcode['places']:
    name.append(place['place name'])
    post.append(place['post code'])
    longitude.append(place['longitude'])
    latitude.append(place['latitude'])
```

```python
print(name)
print(post)
print(longitude)
print(latitude)
```

# User's Todo

```python
response = requests.get("https://jsonplaceholder.typicode.com/todos")
todos = json.loads(response.text)
```

```python
from IPython.display import Image
Image(filename='img/img2.png', width=500, height=500)
```

```python
todos
```

```python
len(todos)
```

```python
userId = []
id = []
title = []
completed = []


for todo in todos:
    userId.append(todo['userId'])
    id.append(todo['id'])
    title.append(todo['title'])
    completed.append(todo['completed'])
```

```python
print(userId)
print(id)
print(title)
print(completed)
```

```python
user = {}

for todo in todos:
    if todo['completed']:
        try:
            user[todo['userId']] += 1
        except KeyError:
            user[todo['userId']] = 1

top_users = sorted(user.items(), key=lambda x: x[1], reverse=True)
```

```python
print(user)
print(top_users)
```

```python
for user, num_complete in top_users:
    print(f"user {user} completed {num_complete} TODOs")
```
